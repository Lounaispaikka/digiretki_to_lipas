```python
TYPE_PROPERTIES = {
"Kalastuskohde piste" : {
    "bool": ["toilet", "pier"]
},
"Uimaranta" : {
    "bool": ["toilet", "sauna", "pier", "changing-rooms", "shower"]
}, 
"Uimapaikka" : {
    "bool": ["toilet", "sauna", "pier", "changing-rooms", "shower"]
},
"Laavu, kota tai kammi" : {
    "bool": ["toilet"]
}, 
"Tupa" : {
    "bool": ["toilet"]
}, 
"Telttailu ja leiriytyminen" : {
    "bool": ["toilet"]
}, 
"Luontotorni": {
    "bool": ["toilet"]
    },
"Tulen/ruuanlaittopaikka": {
    "bool": ["toilet"]
    },

"opastuspiste": {
    "bool": ["toilet", "parkkipaikka"]
},
"Talviuimapaikka" : {
    "bool": ["toilet", "sauna", "pier", "changing-rooms", "shower", "ice-reduction"]
},
"Koiraurheilualue": {
    "bool": ["toilet", "ligthing"]
},
"Kaukalo": {
    "bool": ["toilet", "ligthing", "changing-rooms"]
},
"Ulkoilumaja/hiihtomaja": {
    "bool": ["toilet"]
},
"Veneilyn palvelupaikka": {
    "bool": ["toilet", "pier", "boat-launching-spot"]
},
}
```

```python
KEYWORDS = {
    "laituri": ["laituri", "pier"],
    "sauna": ["sauna"],
    "pukukopit": ["pukuhuone", "changing room", "pukukopit"],
    "suihku": ["suihku", "shower"],
    "yleisowc": ["wc", "kuivakäymälä", "käymälä", "huussi", "puucee", "vessa", "inva-wc", "yleisöwc"],
    "jäätymisenesto": ["jäätymisenesto"],
    "veneen laskupaikka": ["veneenlaskupaikka"]
}
```

```python

import json

def extract_properties(row):
    # Combine equipment, info_fi, and accessibility text
    text = f"{row.get('equipment', '')} {row.get('info_fi', '')} {row.get('accessibil', '')}".lower()
    props = {}

    # Only look up type properties if lipas_type_name is not null
    if pd.notna(row.get('lipas_type_name')):
        type_config = TYPE_PROPERTIES.get(row['lipas_type_name'], {})

        # Booleans
        for prop in type_config.get("bool", []):
            keywords = KEYWORDS.get(prop, [])
            props[prop + "?"] = any(k in text for k in keywords)

        # Categorical
        for prop, values in type_config.get("categorical", {}).items():
            for val in values:
                if val in text:
                    props[prop] = val
                    break

    # Always add accessibility info if available
    if pd.notna(row.get('accessibil')):
        props["accessible?"] = row['accessibility_prediction']  # assuming it's True/False

    # Convert to JSON string for direct export
    return json.dumps(props)
```

```python
import pandas as pd

# Apply the function to each row and store the results in a new column
data_to_export['properties'] = data_to_export.apply(extract_properties, axis=1)

# Check the result
data_to_export.shape
```
