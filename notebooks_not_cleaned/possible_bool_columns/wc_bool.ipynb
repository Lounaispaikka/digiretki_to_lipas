# Boolean column does the location have a toilet

***

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("points_with_accessibility_bool.csv")
```

```python
keywords = ["wc", "kuivakäymälä", "käymälä", "huussi", "puucee", "vessa", "inva-wc"]

def check_toilet(row):
    equip = row['equipment']
    info = row['info_fi']
    
    if pd.isna(equip) and pd.isna(info):
        return np.nan

    combined_text = f"{equip} {info}".lower()
    return any(keyword in combined_text for keyword in keywords)

df['has_toilet'] = df.apply(check_toilet, axis=1)
```
