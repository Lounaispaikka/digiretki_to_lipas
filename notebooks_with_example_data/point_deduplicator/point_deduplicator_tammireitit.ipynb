```python
# imports
import pandas as pd
import numpy as np
import sys
sys.path.append("../../src")

from point_deduplicator import process_data
```

```python
# USER CONFIGURATION 
# change these paths

DATA_PATH = "spatial_distance_search.csv"
MAPPING_PATH = "tammireitit_lipas_type_mapping.csv"

df = pd.read_csv(DATA_PATH)
mapping_df = pd.read_csv(MAPPING_PATH, sep=";")

print("Data preview:")
display(df.head())

print("Mapping preview:")
display(mapping_df.head())

print("\nColumns:")
print(df.columns.tolist())
```

```python
# USER CONFIGURATION
# Add here your column names

ID_COL = "source_id"
LIPAS_ID_COL = "lipas_id"
CLASS_COL = "tyyppi"
TYPE_COL_LIPAS = "lipas_tyyppi_nimi_fi"

MAP_KEY = "tammireitti_tyyppi"
MAP_VALUE = "lipas_type_name"

NAME1_COL = "kohde"
NAME2_COL = "lipas_nimi_fi"

# Optional
DISTANCE = "distance"
```

```python
true_matches, true_matches_duplicates, mismatches, review_mismatches_duplicates, needs_checking, not_matched = process_data(
    df,
    mapping_df,
    id_col=ID_COL,
    lipas_id_col=LIPAS_ID_COL,
    class_col=CLASS_COL,
    type_col=TYPE_COL_LIPAS,
    map_key=MAP_KEY,
    map_value=MAP_VALUE,
    name1_col=NAME1_COL,
    name2_col=NAME2_COL
)
```

```python
# Review that the results are correct

true_matches.loc[:, [ID_COL, LIPAS_ID_COL, CLASS_COL, TYPE_COL_LIPAS, NAME1_COL, NAME2_COL, DISTANCE]]
```

```python
# Manipulate results example
# Row indexes that should NOT be true matches

# Row indexes to be moved
row_idxs_to_move = [0, 1, 2, 6, 7, 8, 9, 10, 12, 13, 15, 17, 18, 19, 24, 30, 31, 33, 34, 35]

# Get the rows from the original dataframe
rows_to_move = true_matches.loc[row_idxs_to_move]

# Concatenate the rows to the correct dataframe
mismatches = pd.concat([mismatches, rows_to_move], ignore_index=True)

# Drop rows from the original dataframe
true_matches = true_matches.drop(index=row_idxs_to_move)
```

```python
# Review duplicate true_matches

true_matches_duplicates.loc[:, [ID_COL, LIPAS_ID_COL, CLASS_COL, TYPE_COL_LIPAS, NAME1_COL, NAME2_COL, DISTANCE]]
```

```python
# Select the correct match from the candidates that have duplicate options

# Row indexes to be moved
row_idxs_to_true = [0, 2] 

# Get the rows from the original dataframe
rows_to_move = true_matches_duplicates.loc[row_idxs_to_true]

# Concatenate the rows to the correct dataframe
true_matches = pd.concat([true_matches, rows_to_move], ignore_index=True)

# Drop from the original dataframe
true_matches_duplicates = true_matches_duplicates.drop(index=row_idxs_to_true)
```

```python
# Review mismatched entries
mismatches.loc[:, [ID_COL, CLASS_COL, TYPE_COL_LIPAS, NAME1_COL, NAME2_COL, DISTANCE]]
```

```python
# Source IDs that should be moved from mismatches to true matches
ids_to_move = [284]

move_mask = mismatches[ID_COL].isin(ids_to_move)

# Move them to true_matches
true_matches = pd.concat(
    [true_matches, mismatches.loc[move_mask]],
    ignore_index=True
)

# Remove them from mismatches
mismatches = mismatches.loc[~move_mask].copy()
```

```python
# Review mismatched duplicate entries

review_mismatches_duplicates.loc[:, [ID_COL, CLASS_COL, TYPE_COL_LIPAS, NAME1_COL, NAME2_COL]]
```

```python
# Move a mismatch with duplicates to true match

# Row indexes to be moved
row_idxs_to_true = [12] 

# Get the rows from the original dataframe
rows_to_move = review_mismatches_duplicates.loc[row_idxs_to_true]

# Concatenate the rows to the correct dataframe
true_matches = pd.concat([true_matches, rows_to_move], ignore_index=True)

# Drop from the original dataframe
review_mismatches_duplicates = review_mismatches_duplicates.drop(index=row_idxs_to_true)
```

```python
# Review results that need a manual check
with pd.option_context("display.max_rows", None):
    display(needs_checking.loc[:,[ID_COL, CLASS_COL, TYPE_COL_LIPAS, NAME1_COL, NAME2_COL, DISTANCE]])
```

```python
# Move rows from needs_checking to true matches

# Row indexes to be moved
row_idxs_to_true = [54, 102, 105, 158, 168, 170]

# Get the rows from the original dataframe
rows_to_move = needs_checking.loc[row_idxs_to_true]

# Concatenate the rows to the correct dataframe
true_matches = pd.concat([true_matches, rows_to_move], ignore_index=True)

# Drop from the original dataframe
needs_checking = needs_checking.drop(index=row_idxs_to_true)
```

```python
# When you are sure of your mismatching rows you can add them to the not_matched df
# to represent that a valid match was not found for the entry

not_matched = pd.concat([not_matched, mismatches], axis=0, ignore_index=True)
not_matched = pd.concat([not_matched, needs_checking], axis=0, ignore_index=True)
not_matched = pd.concat([not_matched, review_mismatches_duplicates], axis=0, ignore_index=True)
not_matched = pd.concat([not_matched, true_matches_duplicates], axis=0, ignore_index=True)
```

```python
# Then you can mark the columns as nan to represent that a match was not found

not_matched[[LIPAS_ID_COL,TYPE_COL_LIPAS, NAME2_COL, 'mapped_lipas_category', 'category_match']] = np.nan
not_matched.head()
```

```python
# After you have reviewed the results and are sure of your matches and mismatches you can make the final matching df

final_matching_df =  pd.concat([not_matched, true_matches], axis=0, ignore_index=True)
final_matching_df
```

```python
# Then you can take a subset of columns of the final matching so it is easy to use with a database table for example
# Here is your dataset identificator column and the matched Lipas entry's identificator column

final_matching_df = final_matching_df[[ID_COL, LIPAS_ID_COL]].copy()
final_matching_df
```

```python
final_matching_df[[ID_COL, LIPAS_ID_COL]] = final_matching_df[[ID_COL, LIPAS_ID_COL]].astype("Int64")
final_matching_df.dtypes
```

```python
# Those can be renamed for clarity if wished

final_matching_df.rename(columns={LIPAS_ID_COL: "mapped_lipas_id", ID_COL: "my_id"}, inplace=True)
final_matching_df
```

```python
# Let's fill the nans as 0 because in our data model 0 represents no match found

final_matching_df["mapped_lipas_id"] = final_matching_df["mapped_lipas_id"].fillna(0)
final_matching_df
```

```python
# Drop extra rows that have resulted from the spatial distance search, these should all be mapped to 0

final_matching_df = final_matching_df.sort_values(by=["my_id", "mapped_lipas_id"],ascending=[True, False])
final_matching_df = final_matching_df.drop_duplicates(subset="my_id",keep="first").reset_index(drop=True)
```

```python
# Sanity check

print(f"Found duplicates: {(final_matching_df["mapped_lipas_id"] > 0).sum()}")
print(f"Number of unique ids: {final_matching_df['my_id'].nunique()}")
print(f"All ids unique: {final_matching_df['my_id'].is_unique}")
```

```python
# Then you can save results to your device if wanted

final_matching_df.to_csv("final_matching_df2.csv", index=False)
```
